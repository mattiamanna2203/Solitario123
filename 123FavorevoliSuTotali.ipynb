## Prova

Prova ridotta per testare il funzionamento della procedura dato che  40! è un numero enorme di permutazioni.   
Ed impiegherebbe troppo tempo per runnare.

```python
carte = [f"{i} {seme}" for seme in ("Spade","Denara","Bastoni","Coppe") for i in range(1,11)]   
numero_permutazioni_possibili = permutazioni_possibili[len(carte)]
print(f"Ci sono {numero_permutazioni_possibili} permutazioni possibili")
count = 1
casi_favorevoli = []

for permutazione in tqdm(itertools.permutations(carte),total=numero_permutazioni_possibili):
   
   permutazione_pulita = tuple(map(lambda carta: int(carta.split()[0]), permutazione))
   insuccessi = np.sum(sequenza_sconfitta == permutazione_pulita)
   if insuccessi == 0:
      casi_favorevoli.append(permutazione)
   count += 1
casi_favorevoli
```
